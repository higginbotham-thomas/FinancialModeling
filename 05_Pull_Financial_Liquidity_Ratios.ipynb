```python
import config
import functions
from functions import get_jsonparsed_data
```

```python
#Financial Modeling Prep URL
url = 'https://financialmodelingprep.com/api/v3/ratios/AAPL?limit=40&apikey='+config.FMPKEY
```

```python
Ratios_AAPL = get_jsonparsed_data(url)
```

```python
#show the parsed data that was returned
Ratios_AAPL
```

```python
# Shows Apple's 2022 Cash Ratio
Ratios_AAPL[0]['cashRatio']
```

```python
# Shows the dates for each index
for i in range(0,len(Ratios_AAPL)):
    print(i, Ratios_AAPL[i]['date'])
```

```python
#This block of code will print out all of the liquidity ratios for a certain year
#
#   Current Ratio
#   Quick Ratio
#   Cash Ratio
#   Days of Sales Outstanding (DSO)
#   Days of Inventory Outstanding (DIO)
#   Operating Cycle
#   Days of Payables Outstanding (DPO)
#   Cash Conversion Cycle

LiquidityRatios = [
        'currentRatio',
        'quickRatio',
        'cashRatio',
        'daysOfSalesOutstanding',
        'daysOfInventoryOutstanding',
        'operatingCycle',
        'daysOfPayablesOutstanding',
        'cashConversionCycle',
    ]
#This is a list of all the profitability ratios 

AAPL_2022_Ratios = []
#This is an empty list of that the next for loop will fill

for i in range(0,len(LiquidityRatios)):
    AAPL_2022_Ratios.append(Ratios_AAPL[0][LiquidityRatios[i]])
    #This adds the ratio i to the list of Apple's ratios

    print(f"{LiquidityRatios[i]} , {AAPL_2022_Ratios[i]: .2f}")

#Notice in the last line code within the f string there is a “: .2f”. This little addition formats the output of the print function to be to two decimal places. This makes the output far more digestible.
```

```python
#This next block of code will print out the cashRatio for the past 5 years:

for i in range(0,5):
    print(f"Cash Ratio = {Ratios_AAPL[i]['cashRatio']:.2f}")
```

```python
# Finally, here’s a function you can use to get a print out of any ratios historical values:

def historical_ratio(ratioName, years):
    for i in range(0,years):
        print(f"ratioName = {Ratios_AAPL[i][ratioName]:.2f}")
```

```python
historical_ratio('quickRatio',5)
```
